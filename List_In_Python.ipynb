# Important List Methods

```python
#List Creation
numbers = [1, 2, 3, 4, 5]
names = ["Kartik", "Rahul", "Amit"]
mixed = [1, "Hello", 3.5, True]

print(numbers)
print(names)
print(mixed)
```

```python
#Accessing Elements
print(numbers[0])    # First element --> Basic Iteration Purpose usage
print(numbers[4])  # Last element 
print(numbers[-1])   # Last element --> Mostly useful in string reverse
```

```python
#Slicing the Lists --> [<start_index>:<excluded_end_index>:<optional_gap>]
print(numbers[1:4])   # [2, 3, 4]
print(numbers[:3])    # [1, 2, 3]
print(numbers[::2])   # [1, 3, 5] --> Double Colone means --> 0 : (n-1) : maintain gap 2
print(numbers[::-2])   # Reverse Operation --> 0 : (n-1) : maintain -2 gap represents in reverse operation 
```

```python
#Adding elements
numbers.append(6)          # Add at end --> Add an element
numbers.insert(2, 100)     # Insert at index ---> Insert at specific index value
numbers.extend([7, 8])     # Add multiple elements ---> add another list with the list

print(numbers)
```

```python
#Removing Elements
numbers.remove(100)   # Remove specific value --> **not return** the element
numbers.pop()         # Remove last element --> just like stack --> **not return** the element
numbers.pop(2)        # Remove by index ---> **return** the element
```

```python
print(numbers)
```

```python
#Updating Elements
numbers[0] = 10
print(numbers)
```

```python
#length of List
print(len(numbers))
```

```python
#loop through list
#Method_1:
for item in numbers:  #item --> acts as a value
    print(item,end=" ")

print("\n")
#Method_2:
for item in range (0,len(numbers)): #item --> acts as an index
    print(numbers[item],end=" ") 
```

```python
#Check/Find elements
if 5 in numbers:
    print("Found")
```

```python
#Sorting by 'sort'

numbers.sort()              
print("Ascending:",numbers)

numbers.sort(reverse=True) 
print("Descending:", numbers)
```

```python
#Reverse Operation
numbers.reverse()
print(numbers)
```

```python
#Copy List
copy_list = numbers.copy()
print(copy_list)
```

```python
#List Comprehension
squares = [x*x for x in range(6)]
print(squares)
```

```python
#nested list
matrix = [
    [1, 2],
    [3, 4]
]

print(matrix[0][1])   # 2
```

```python
#Common elements between two list
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

common = list(set(a) & set(b))

print(common)
```

```python
#merge two lists
a = [1, 2, 3]
b = [4, 5, 6]

merged = a + b

print(merged)
```

```python
#Palindrome List
nums = [1, 2, 3, 2, 1]

if nums == nums[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# Applicataion Coding Based On List

```python
listNum = [12,15,17,24,45,12,36,15,17,25,98,37,65,85,67,34,45,98,100,65,75,65,45,36,25,85,45,15]
```

```python
print(listNum)
```

```python
print(max(listNum))
```

```python
print(min(listNum))
```

```python
print(len(listNum))
```

```python
#Reverse the List
```

```python
revList1 = listNum[::-1] #method 1
print(revList1)
```

```python
print("Actual List:",listNum)
revList2 = []
end_index = len(listNum) - 1
start_index = -1

for i in range (end_index,start_index,-1): #method_2
    revList2.append(listNum[i])
    
print("Reverse List:",revList2)
```

```python
#Frequency of elements
```

```python
freq = {} #dictionary
for i in listNum:
    freq[i] = freq.get(i, 0) + 1

print(freq)
```

```python
#Find Duplicates in the list (standard level)
```

```python
seen = set() #tuple
duplicates = [] #list

for i in listNum:
    if i in seen:
        duplicates.append(i)
    else:
        seen.add(i)
        
print(duplicates)
```

```python
#Find Duplicates with Frequency in the list (advanced level)
```

```python
# Step 1: previous case we find the duplictes as well as frequency; combine these with condition freq>1 to get answer

print("Frequency All:",freq)
print("Duplicates Only:",duplicates)
duplicate_frequencies = {}

# Keep only the items that appeared more than once
for key,value in freq.items(): #.items() --> help to call dictionary items
    if value > 1:
        duplicate_frequencies[key] = value

print("Duplicates with Freq:",duplicate_frequencies)
```

```python
#Linear Search
```

```python
target = 100

for i in range(len(listNum)):
    if listNum[i] == target:
        print("Found at index", i) 
```

```python
#Binary Search
```

```python
target = 98

low = 0
high = len(listNum) - 1

while low <= high:
    mid = (low + high) // 2

    if listNum[mid] == target:
        print("Found")
        break
    elif listNum[mid] < target:
        low = mid + 1
    else:
        high = mid - 1
```
